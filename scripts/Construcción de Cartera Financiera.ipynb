### Construcción de la Clase

```python
class fwallet: #Al inicializar la clase, lee inputs registrados, lee bases de datos y los guarda como valores propios

    def __init__(self,inputs): #Función de bienvenida
        self.inputs = inputs #Vinculación con una base de datos JSON (Utilizar diccionarios para crear bases de datos, buscar:Read JSON Python)
                            #Aplicación Principal sirve para vincularse con API de internet
        self.incomes = pd.read_csv(self.inputs['data_route']+self.inputs['incomes_filename'])
        self.expenses = pd.read_csv(self.inputs['data_route']+self.inputs['expenses_filename'])

    ### ACTION TO PERFORM ###

    #Add incomes to database
    def add_incomes(self):
        date = input('date: ')
        amount = input('amount: ')
        category = input('category: ')
        third_party = input('third party: ')
        description = input('description: ')

        new_income = [date,amount,category,third_party,description] #Creación de lista

        n = len(self.incomes)
        self.incomes.loc[n] = new_income

        self.incomes.to_csv(self.inputs['data_route']+self.inputs['incomes_filename'],index=False)
        
    #Add expenses to database
    def add_expense(self):
        date = input('date: ')
        amount = input('amount: ')
        category = input('category: ')
        third_party = input('third party: ')
        description = input('description: ')

        new_expense = [date,amount,category,third_party,description] #Creación de lista

        n = len(self.expenses)
        self.expenses.loc[n] = new_expense

        self.expenses.to_csv(self.inputs['data_route']+self.inputs['expenses_filename'],index=False)
    
    #Calculate the total incomes
    def total_income(self): #Función que muestre el ingreso total
        total = self.incomes['AMOUNT'].sum()
        return total
        
    #Calculate the total expenses
    def total_expenses(self): #Función que muestre el gasto total
        total = self.expenses['AMOUNT'].sum()
        return total
    
    #calculate balance = income - expenses
    def balance(self): #Función que muestre el balance o diferencia entre ingresos y balance
        balance = self.total_income() - self.total_expenses()
        return balance

    #Create financial reporte usinf the functions above
    def financial_report(self): #Función que muestre un reporte financiero
        print('\n'+'Financial Report'+'\n')
        print(f'Total Income: ${self.total_income()}')
        print(f'Total expense: ${self.total_expenses()}')
        print(f'Balance: ${self.balance()}')
        if self.balance() > 0:
            print('Don´t forget to save some of your left-off money! Great job!')
        else:
            print('Be responsible with your finance, and find a low rate loan!')
```

```python
#Crear variable inputs como diccionario (Truco de un Quant, recomendable volverse Quant)
inputs = {
    'data_route' : '../data/',
    'expenses_filename' : 'expenses.csv',
    'incomes_filename' : 'incomes.csv'
    
}
```

```python
cartera = fwallet(inputs) #Creación de variable y se instancia de fwallet
```
